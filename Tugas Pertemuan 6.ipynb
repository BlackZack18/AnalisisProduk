## Visualization & Explanatory Analysis

### Pertanyaan 1: 

```python
# Membuat Tabel Tinggi Produk
height_table = df[['product_category_name', 'product_height_cm']]
height_table = height_table.rename(columns={'product_category_name': 'Nama Produk', 'product_height_cm': 'Tinggi Produk (cm)'})

# Informasi Rata-rata, nilai max, dan nilai min
average_height = df['product_height_cm'].mean()
max_height = df['product_height_cm'].max()
min_height = df['product_height_cm'].min()
```

```python
# Tabel dan Informasi
display(height_table)
print(f"Rata rata tinggi produk : {average_height:.2f} cm")
print(f"Produk paling tinggi : {max_height} cm")
print(f"Produk paling pendek : {min_height} cm")
```

```python
# Visualisasi distribusi tinggi produk dan informasinya
plt.figure(figsize=(8, 5))
plt.hist(df['product_height_cm'], bins=20, color='cyan', edgecolor='black')
plt.axvline(x=df['product_height_cm'].mean(), color='red', linestyle='dashed', linewidth=2, label='Rata-rata Tinggi')
plt.title('Distribusi Tinggi Produk dan Garis Referensi')
plt.xlabel('Tinggi Produk (cm)')
plt.ylabel('Frekuensi')

plt.legend()
plt.show()
```

### Pertanyaan 2: 

```python
df['product_weight_kg'] = df['product_weight_g'] / 1000
weight_table = df[['product_category_name', 'product_weight_kg']]
weight_table = weight_table.rename(columns={'product_category_name': 'Nama Produk', 'product_weight_kg': 'Berat Produk (kg)'})
weight_table
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df['product_weight_kg'], bins=15, color='blue', edgecolor='black')
plt.title('Berat Produk')
plt.xlabel('Berat Produk (kg)')
plt.ylabel('Frekuensi')
plt.show()
```

```python
print("Berat Produk Terendah : ", df['product_weight_kg'].min())
print("Berat Produk Tertinggi : ", df['product_weight_kg'].max())
print("Berat Produk Tertinggi : ", df['product_weight_kg'].mean())
```

### Pertanyaan 3: 

```python
length_table = df[['product_category_name', 'product_length_cm']]
length_table = length_table.rename(columns={'product_category_name': 'Nama Produk', 'product_length_cm': 'Panjang Produk (cm)'})
length_table
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df['product_length_cm'], bins=20, color='yellow', edgecolor='black')
plt.title('Panjang Produk')
plt.xlabel('Panjang Produk (cm)')
plt.ylabel('Frekuensi')
plt.show()
```

```python
print("Panjang Produk Terendah : ", df['product_length_cm'].min())
print("Panjang Produk Tertinggi : ", df['product_length_cm'].max())
print("Panjang Produk Rata-rata : ", df['product_length_cm'].mean())
```

### Pertanyaan 4: 

```python
width_table = df[['product_category_name', 'product_width_cm']]
width_table = width_table.rename(columns={'product_category_name': 'Nama Produk', 'product_width_cm': 'Lebar Produk (cm)'})
width_table
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df['product_width_cm'], bins=20, color='pink', edgecolor='black')
plt.title('Diagram Lebar Produk')
plt.xlabel('Lebar Produk (cm)')
plt.ylabel('Frekuensi')
plt.show()
```

```python
print("Produk Paling Lebar    : ", df['product_category_name'].max(), "->", df['product_width_cm'].max(), "cm")
print("Produk Paling Kecil    : ", df['product_category_name'].min(), "->", df['product_width_cm'].min(), "cm")  
print("Lebar Produk Rata-Rata : ", df['product_width_cm'].mean(), "cm")
```

### Pertanyaan 5: 

```python
name_table = df[['product_category_name', 'product_name_lenght']]
name_table = name_table.rename(columns={'product_category_name': 'Nama Produk', 'product_name_lenght': 'Panjang Karakter'})

average_name = df['product_name_lenght'].mean()
max_name = df['product_name_lenght'].max()
min_name = df['product_name_lenght'].min()

display(name_table)
print("Produk Dengan Nama Paling Panjang  : ", df['product_name_lenght'].max())
print("Produk Dengan Nama Paling Pendek   : ", df['product_name_lenght'].min())  
print("Panjang Karakter Rata-Rata : ", df['product_name_lenght'].mean(),)
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df['product_name_lenght'], bins=20, color='blue', edgecolor='black')
plt.title('Grafik Panjang Karakter')
plt.xlabel('Jumlah Karakter')
plt.ylabel('Frekuensi')
plt.show()
```
